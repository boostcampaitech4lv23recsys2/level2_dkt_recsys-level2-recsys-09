```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

import warnings
warnings.filterwarnings("ignore")
```

```python
mpl.style.use('ggplot')
```

```python
def performance_graph(public: list, private: list, title: str):
    x = np.arange(len(public))
    public_best_idx = np.argmax(public)
    private_best_idx = np.argmax(private)
    plt.title(title)
    plt.plot(x, public, "-o", label='public score')
    plt.plot(x, private, "-o", label='private score')
    plt.plot(public_best_idx, public[public_best_idx], "o", label="public best")
    plt.plot(private_best_idx, private[private_best_idx], "o", label="private best")
    plt.legend()
    plt.show()
```

## LightGCN(11-18~11-22)

```python
lgcn_public = [0.7049, 0.7603, 0.7524, 0.7862, 0.7820, 0.7934, 0.7910, 0.7784, 0.7867, 0.7902,]
lgcn_private = [0.7118, 0.7927, 0.7847, 0.8205, 0.8144, 0.8089, 0.8040, 0.8195, 0.8201, 0.8054,]

performance_graph(lgcn_public, lgcn_private, "LightGCN(11-18~11-22)")
```

## TabNet(11-30~12-05)

```python
tab_public = [0.7742, 0.7622, 0.7648, 0.7891, 0.7879, 0.7715, 0.7819, 0.7856, 0.7712, 0.7870,]
tab_private = [0.8226, 0.8166, 0.7938, 0.7997, 0.8005, 0.7703, 0.7997, 0.8087, 0.7940, 0.7965,]

performance_graph(tab_public, tab_private, "TabNet(11-30~12-05)")
```
