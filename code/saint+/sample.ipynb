```python
import time
import pickle
import numpy as np
import pandas as pd
from datetime import datetime

train=pd.read_csv('/opt/ml/input/DKT/data/train_data.csv')
test=pd.read_csv('/opt/ml/input/DKT/data/test_data.csv')
```

```python
avg_item = train.groupby('assessmentItemID').mean().reset_index()[['assessmentItemID','answerCode']]
avg_user = train.groupby('userID').mean().reset_index()[['userID','answerCode']]
train = pd.merge(train, avg_item, on='assessmentItemID', how='inner')
train = pd.merge(train, avg_user, on='userID', how='inner')
train.columns = ['userID','assessmentItemID','testId','answerCode','Timestamp','KnowledgeTag','avg_item','avg_user']

avg_item = test.groupby('assessmentItemID').mean().reset_index()[['assessmentItemID','answerCode']]
avg_user = test.groupby('userID').mean().reset_index()[['userID','answerCode']]
test = pd.merge(test, avg_item, on='assessmentItemID', how='inner')
test = pd.merge(test, avg_user, on='userID', how='inner')
test.columns = ['userID','assessmentItemID','testId','answerCode','Timestamp','KnowledgeTag','avg_item','avg_user']
```
